## 2. Reporting dynamique (DASH)

```python
# Chemin du fichier Excel
file_path = r"C:/Users/gaeta/Documents/03_ PYTHON/14_PROJET_PCM/02_INPUTS/_20230301_inputs_projet_pcm.xlsx"
print(file_path)
```

```python
# Vérification de l'existence du fichier
if not os.path.exists(file_path):
    raise FileNotFoundError(f"Le fichier {file_path} est introuvable !")
```

```python
# Chargement des données depuis l'Excel
df = pd.read_excel(file_path,sheet_name="CYCLISTS")
print(df.shape)
df.head()
```

```python
df=df[['IDcyclist','Nom','Prenom','ID_team','fklDregion','Date_de_naissance','Popularite','value_f_potentiel','taille_coureur','poids_coureur','carac_plaine','carac_montagne','carac_descente','carac_paves','carac_clm','carac_prologue','carac_sprint','carac_acceleration','carac_endurance','carac_resistance','carac_recuperation','carac_vallon','carac_baroudeur','prendra_sa_retraite','Coureur_champion','gene_ilist_flkDfavorite_races','value_i_yearneopro','gene_i_nb_victory','gene_i_nb_tdf','gene_i_nb_giro','gene_i_nb_vuelta','gene_i_nb_sanremo','gene_i_nb_flandres','gene_i_nb_roubaix','gene_i_nb_liege','gene_i_nb_lombardia']]

# Calcul du niveau moyen par coureur
df["carac_moy"]=round(((df['carac_plaine'] + df['carac_montagne'] + df['carac_descente'] +
df['carac_paves'] + df['carac_clm'] + df['carac_prologue'] + df['carac_sprint'] +
df['carac_acceleration'] + df['carac_endurance'] + df['carac_resistance'] + 
df['carac_recuperation'] + df['carac_vallon'] + df['carac_baroudeur'])/(13)),2)

print(df.shape)
df.head()
```

```python
# Création de l'application Dash
app = dash.Dash(__name__)

# Mise en page de l'application
app.layout = html.Div([
    html.H1("Tableau interactif - Reporting Projet PCM"),
    
    # Affichage du tableau interactif
    dash_table.DataTable(
        id='table-data',
        data=df.to_dict('records'),
        columns=[{"name": col, "id": col} for col in df.columns],
        page_size=10,  # Nombre de lignes affichées par page
        sort_action="native",  # Activation du tri
        filter_action="native",  # Activation du filtre
        style_table={'overflowX': 'auto'},
        style_header={'backgroundColor': 'lightgrey', 'fontWeight': 'bold'},
        style_cell={'textAlign': 'left'}
    )
])
```

```python
# Exécution de l'application
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import streamlit as st
import pandas as pd

# Chemin du fichier Excel
excel_path = r"C:/Users/gaeta/Documents/03_ PYTHON/14_PROJET_PCM/02_INPUTS/_20230301_inputs_projet_pcm.xlsx"

# Charger le fichier Excel
@st.cache_data
def load_data():
    return pd.read_excel(excel_path, engine="openpyxl")

df = load_data()

# Titre de l'application
st.title("Analyse des données Excel avec Streamlit")

# Ajouter une barre de recherche
search_text = st.text_input("Rechercher dans le DataFrame")

# Filtrer les données si une recherche est effectuée
if search_text:
    df = df[df.apply(lambda row: row.astype(str).str.contains(search_text, case=False).any(), axis=1)]

# Afficher le DataFrame interactif
st.dataframe(df, use_container_width=True)

# Ajouter un graphique interactif (exemple)
if not df.empty:
    st.subheader("Graphique interactif")
    numeric_columns = df.select_dtypes(include=["number"]).columns.tolist()

    if len(numeric_columns) >= 2:
        x_axis = st.selectbox("Choisir l'axe X", numeric_columns, index=0)
        y_axis = st.selectbox("Choisir l'axe Y", numeric_columns, index=1)
        st.line_chart(df[[x_axis, y_axis]])
    else:
        st.warning("Pas assez de colonnes numériques pour afficher un graphique.")
```
